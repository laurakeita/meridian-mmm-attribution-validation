# 04 · Budget Optimization
Optimize media spend allocation using Meridian's built-in optimizer.

*Assumes `mmm`, `data`, and `media` are already in memory from previous notebooks.*

```python
from meridian.analysis import optimizer
import IPython
```

## Optimization Setup
- Budget = total actual spend in the full data period
- Constraints: each channel can vary between 50% and 150% of its current spend

```python
start_date_opt = data['date'].min().strftime('%Y-%m-%d')
end_date_opt   = data['date'].max().strftime('%Y-%m-%d')

opt_data = data[(data['date'] >= start_date_opt) & (data['date'] <= end_date_opt)]
current_spend = opt_data[media].sum()
total_budget  = int(current_spend.sum())
n_channels    = len(media)

spend_constraint_lower = [0.5] * n_channels
spend_constraint_upper = [1.5] * n_channels

print(f'Optimization period: {start_date_opt} → {end_date_opt}')
print(f'Total budget: {total_budget:,}')
print(f'Channels ({n_channels}):', media)
print('\nCurrent spend per channel:')
print(current_spend)
```

## Run Optimization

```python
budget_optimizer = optimizer.BudgetOptimizer(mmm)

optimization_results = budget_optimizer.optimize(
    start_date=start_date_opt,
    end_date=end_date_opt,
    budget=total_budget,
    spend_constraint_lower=spend_constraint_lower,
    spend_constraint_upper=spend_constraint_upper,
)
```
